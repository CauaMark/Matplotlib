```python
import matplotlib
import matplotlib.pyplot as plt

# Comando plot é usado para desenhasr os marcadores do gráfico
plt.plot([1,2,3,4], [1,4,9,16]) #(x, y)
plt.show()
```

```python
# Importando numpy e já utilizando a abreviação np, pode se usar as listas no plot:
import matplotlib.pyplot as plt
import numpy as np
eixoX = np.array([1, 8])
eixoY = np.array([3, 10])
plt.plot(eixoX, eixoY)
plt.show()
```

```python
# subplot: Com o comando subplot, pode ser desenhado mais de um gráfico na mesma imagem. 
import matplotlib.pyplot as plt
import numpy as np
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1) #(quantidade de linhas , quantidade de colunas,)
plt.plot(x,y)
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.show()
```

```python
# Bar: É possível criar gráfico de barra usando o comando bar: plt.bar(x,y)
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y)
plt.show()
```

```python
# Scatter: Com o comando scatter o gráfico gerado é um gráfico de dispersão: 
plt.scatter(x, y)
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)
plt.show()
```

```python
# hist: o comando hist cria um histograma:
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

```python
# pie: comando pie cria uma gráfico de pizza:
import matplotlib.pyplot as plt
y =[35, 25, 25, 15] # a soma dos valores deve ser 100
plt.pie(y)
plt.show()
```

```python
# label: comando label é usado para criar títulos ou legendas para as fatias.
import matplotlib.pyplot as plt
y = [35, 25, 25, 15]
mylabels =["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.show()
```

```python
# explode:  Este comando só é usado para gráficos do tipo pizza.Ele serve para destacar uma porção no gráfico de pizza
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0] # Aqui se define a distancia que irá ser destacada
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
# shadow() : Este comando coloca uma sombra no gráfico de pizza
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True) # Ele deve ser colocado junto ao comando que cria o gráfico e junto a ele deve se escrever "True"
plt.show()
```

```python
# plt.title() = Este comando coloca um título no gráfico, o título deve ser colocado dentro dos parênteses e entre aspas.
#                      "plt.title("Título do gráfico")"
# xlabel() e ylabel: esses comandos são usados para definir um título para os eixos x e y.
import numpy as np
import matplotlib.pyplot as plt
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Título do gráfico")
plt.show()
```

```python
# plt.suptitle() = Este comando coloca um super título, é usado quando deseja mostrar um título mais importante.
# Ex: plt.suptitle("Minha loja")
import matplotlib.pyplot as plt
import numpy as np
#plot 1:
plt.suptitle("Minha loja")
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Vendas")
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Encomendas")
plt.show()
```

```python
# plt.legend() = Esta função mostra uma legenda sobre o gráfico no gráfico, ela pode ser colocada em um lugar específico do gráfico.A legenda se não for indicada a posição, ela vem como padrão no canto superior esquerdo.

# Ex1
import matplotlib.pyplot as plt

y1 = [2, 3, 4.5]
y2 = [1, 1.5, 5]
plt.subplot(1, 1, 1)
plt.plot(y1)
plt.plot(y2)
plt.legend(["Azul", "Laranja"], loc="lower right") # Aqui com o "loc" voce define em qual lugar a legenda deve aparecer
plt.show()

# Ex2 apenas uma leganda
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.legend(['Legenda'])
plt.show()
```

```python
# plt.figure(figsize=()) : Esse comando aumenta ou diminui o dashbord que o gráfico é mostrado
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.figure(figsize=(20, 20)) # Se difine o eixo x e o eixo y dentro dos parenteses
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.legend(['Legenda'])
plt.show()
```

```python
#Código de médias escolares
import matplotlib.pyplot as plt
import numpy as np
print(30*"=")
print("GRÁFICOS DE MÉDIAS DO SEMESTRE")
print(30*"=")
nomes = []
notas = []
vezes = int(input("Digite quantas médias serão lançadas: "))
for i in range(vezes):
    nomes.append(input("Digite o nome do aluno(a): "))
    notas.append(float(input("Digite a média: ")))
if len(notas) == len(nomes):
    x = np.arange(len(notas))
    colors=[]
    for i in range(len(notas)):
        if notas[i]>=7:
            colors.insert(i, "green")
        else:
            colors.insert(i, "red")
    fig, ax = plt.subplots()
    ax.bar(x, notas, alpha=0.5, color=colors)
    plt.axhline(y = 7, color = 'k', linestyle = '-')
    plt.suptitle("Média dos alunos")
    plt.title("Média baseada nas notas do decorrente semestre")
    plt.xticks(x, nomes)
    plt.show()
else:
    print("Verifique os seus dados")
```

```python
#Código de animação de linha
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
fig, ax = plt.subplots(figsize=(5, 3))
ax.set(xlim=(-3, 3), ylim=(-1, 1))
x = np.linspace(-3, 3, 91)
t = np.linspace(1, 25, 30)
X2, T2 = np.meshgrid(x, t)
sinT2 = np.sin(2*np.pi*T2/T2.max())
F = 0.9*sinT2*np.sinc(X2*(1 + sinT2))
line = ax.plot(x, F[0, :], color='k', lw=2)[0]

def animate(i):
    line.set_ydata(F[i, :])
anim = FuncAnimation(
fig, animate, interval=100, frames=len(t)-1)
plt.draw()
plt.show()
```
